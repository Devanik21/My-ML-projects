```python
from keras.layers import LeakyReLU

# Standardize the data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Build the ANN model
model = Sequential()
model.add(Dense(units=256, input_dim=4))
model.add(LeakyReLU(alpha=0.1))
model.add(Dense(units=128))
model.add(LeakyReLU(alpha=0.1))
model.add(Dense(units=64))
model.add(LeakyReLU(alpha=0.1))
model.add(Dense(units=8))
model.add(LeakyReLU(alpha=0.1))
model.add(Dense(units=2))
model.add(LeakyReLU(alpha=0.1))
model.add(Dense(units=1, activation='sigmoid'))

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(X_train_scaled, y_train, epochs=20, batch_size=1, validation_data=(X_test_scaled, y_test))
```

```python
def predict_output_binary(X_series):
    X_series_scaled = scaler.transform(X_series)
    predictions = model.predict(X_series_scaled)
    binary_predictions = np.round(predictions).astype(int)
    return binary_predictions.flatten()

# Example input series (replace with your own series)
X_series_example = np.array([[1,3,5,7]])

# Predict output for the example input series in binary
predictions_example_binary = predict_output_binary(X_series_example)
print("Predicted output for the example input series (in binary):")
print(predictions_example_binary)
```
