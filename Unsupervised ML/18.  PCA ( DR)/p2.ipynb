4D to 2D

Using PCA class

```python
from sklearn.decomposition import PCA
p=PCA(n_components=2)
```

```python
projected_data=p.fit_transform(x_scaled)
projected_data
```

```python
pc1=projected_data[: , 0]
pc2=projected_data[: , 1]
```

   A new way to create a dataframe

```python
df1 = DataFrame({'pc1': pc1, 'pc2': pc2, 'target': y}) 
df1
```

```python
pairplot(hue='target',data=df1)
```

Now, train the model

```python
df1=df1.drop('target', axis=1)
```

```python
df1
```

Now, let's feed to the ML model 

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df1,y,test_size=0.2,random_state=100)
```

```python
from sklearn.linear_model import  LogisticRegression
l=LogisticRegression()    # or SVC
l.fit(x_train,y_train)
```

```python
l.score(x_train,y_train)
```

```python
l.score(x_test,y_test)
```
