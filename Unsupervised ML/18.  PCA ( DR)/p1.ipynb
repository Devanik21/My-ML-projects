4D to 1D

```python

from pandas import  *
from numpy import *
from seaborn import *
from matplotlib.pyplot import *
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
```

```python
from sklearn.datasets import  load_iris
iris=load_iris()
iris
```

```python
iris.data
```

```python
iris.feature_names
```

```python
x=DataFrame(iris.data,columns=iris.feature_names)
```

```python
x
```

```python
y=iris.target
y
```

Dimensionality reduction by standard scaler

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
x_scaled=scaler.fit_transform(x)
```

```python
x_scaled
```

create covariance matrix

```python
rows=x_scaled.T
```

```python
cov_matrix=cov(rows)
cov_matrix
```

Eigen values & Eigen vectors

```python
eig_vals,eig_vecs=linalg.eig(cov_matrix)
```

```python
eig_vals
```

```python
eig_vecs
```

The below code gives percentage of each Eigen value

```python
for i in range(len(eig_vals)):
    print(eig_vals[i]/sum(eig_vals))
```

max. eigen value is in 0th row

```python
eig_vecs.T
eig_vecs.T[0]
```

```python
projected_x=x_scaled.dot(eig_vecs.T[0])
projected_x
```

```python
df=DataFrame(projected_x,columns=['PC1'])
df['y-axis']=0.0
df['target']=y
df
```

```python
pairplot(data=df,hue='target')
```

```python
x_scaled.shape
```

```python
projected_x.shape
```

see we converted 4D data into 1D data

2nd method - PCA class....................(easy)

```python
from sklearn.decomposition import PCA
p=PCA(n_components=1)
```

```python
projected_x=p.fit_transform(x_scaled)
projected_x
```

```python
df1=DataFrame(projected_x,columns=['PC1'])
df1['y-axis(PC2)']=0.0
df1['target']=y
df1
```

Now use the reduced data to feed on ML model

```python
df_reshaped=projected_x.reshape(-1,1) #convert into 2D array
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df_reshaped,y,test_size=0.2,random_state=100)
```

```python
from sklearn.linear_model import  LogisticRegression
l=LogisticRegression()      # or SVC
l.fit(x_train,y_train)
```

```python
l.score(x_train,y_train)
```

```python
l.score(x_test,y_test)
```
