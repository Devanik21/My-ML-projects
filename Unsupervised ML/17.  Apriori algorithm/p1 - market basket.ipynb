```python
from apyori import apriori

# Assuming lst is your transaction data
f = open('Market_Basket_Optimisation.csv', 'r')
transactions = [line.strip().replace('"', '').split(',') for line in f]
f.close()

# Perform Apriori algorithm
result_list = list(apriori(transactions, min_support=0.003, min_confidence=0.4, min_lift=4, min_length=2))

# Print or use the results as needed
for record in result_list:
    items = list(record.items)
    support = record.support
    ordered_statistics = record.ordered_statistics[0]

    items_base = list(ordered_statistics.items_base)
    items_add = list(ordered_statistics.items_add)
    confidence = ordered_statistics.confidence
    lift = ordered_statistics.lift

    print(f"Items: {items}")
    print(f"Support: {support}")
    print("Ordered Statistics:")
    print(f"  Items Base: {items_base}")
    print(f"  Items Add: {items_add}")
    print(f"  Confidence: {confidence}")
    print(f"  Lift: {lift}\n")
```
